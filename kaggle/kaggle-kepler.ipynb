```python
import numpy as np 
import pandas as pd 
import os 
def load_kaggle_data(filename, dataset_slug):
    """
    Automatically detects the environment and loads a CSV file 
    using either the local path or the Kaggle server path.
    
    Parameters:
    - filename (str): The name of the file (e.g., 'patients.csv')
    - dataset_slug (str): The dataset folder name used on Kaggle (e.g., 'username/dataset_name')
    """

    kaggle_path = os.path.join('/kaggle/input/datasets/', dataset_slug, filename)
  
    if os.path.exists(kaggle_path):
        print(f"✈️ Running on Kaggle. Loading: {kaggle_path}")
        return pd.read_csv(kaggle_path)
    else:
        print(f"🏠 Running locally. Loading: {filename}")
        return pd.read_csv('../data/'+filename)
```

```python
kepler_data = load_kaggle_data("cumulative.csv","nasa/kepler-exoplanet-search-results")
```

```python
kepler_data = kepler_data.drop(["kepler_name","koi_teq_err2","koi_teq_err1","koi_score"],axis=1)
kepler_data
```

```python
kepler_data.dropna(inplace=True)
kepler_data.shape
```

```python
# Create a binary target column
kepler_data["target"] = (kepler_data["koi_disposition"] == "CONFIRMED").astype(int)

# Now 'target' will be 1 for CONFIRMED, and 0 for both CANDIDATE and FALSE POSITIVE

kepler_data = kepler_data.drop(["rowid","kepid","koi_pdisposition",],axis=1)
kepler_data
```

```python
final_data = kepler_data[kepler_data["koi_disposition"].isin(["CONFIRMED", "FALSE POSITIVE"])].copy()
# Remove the flags entirely! 

# Force the model to only use light curve and stellar properties
scientific_features = [
    "koi_period",    # Orbital period (days)
    "koi_duration",  # Transit duration (hours)
    "koi_depth",     # Transit depth (parts per million)
    "koi_prad",      # Planetary radius (Earth radii)
    "koi_teq",       # Equilibrium temperature (Kelvin)
    "koi_insol",     # Insolation flux (Earth flux)
    "koi_steff",     # Stellar effective temperature (Kelvin)
    "koi_srad"       # Stellar radius (Solar radii)
]

# Drop rows with missing values
clean_data = final_data[scientific_features + ["target"]].dropna()

X = clean_data[scientific_features]
y = clean_data["target"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Initialize and train the model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate!
print(f"Model Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print(classification_report(y_test, y_pred))
```
