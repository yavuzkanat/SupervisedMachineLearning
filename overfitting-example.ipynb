```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
import pandas as pd
```

```python
y_df = pd.DataFrame(data.target)
X_df = pd.DataFrame(data.data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))

plt.scatter(X_df.iloc[:10,0],y_df.iloc[:10],color="r")

plt.title("0. Feature")

plt.grid(linestyle="--")

plt.xlabel("0. Feature (First 10 Samples)")
plt.ylabel("Target y (First 10 Samples)")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_df,y_df,random_state=42,test_size=0.3)
```

```python
dt_overfitting_model = DecisionTreeClassifier(random_state=42)

dt_overfitting_model.fit(X_train,y_train)

print(f"Train Score: {dt_overfitting_model.score(X_train, y_train):.4f}")
print(f"Test Score: {dt_overfitting_model.score(X_test, y_test):.4f}")
```

```python
dt_fixed_model = DecisionTreeClassifier(random_state=42,max_depth=3)

dt_fixed_model.fit(X_train,y_train)

print(f"Train Score: {dt_fixed_model.score(X_train, y_train):.4f}")
print(f"Test Score: {dt_fixed_model.score(X_test, y_test):.4f}")
```

```python
dt_fixed_model2 = DecisionTreeClassifier(random_state=42,max_depth=5)

dt_fixed_model2.fit(X_train,y_train)

print(f"Train Score: {dt_fixed_model2.score(X_train, y_train):.4f}")
print(f"Test Score: {dt_fixed_model2.score(X_test, y_test):.4f}")
```

```python
te_scores = []
tr_scores = []

plt.figure(figsize=(8,5))

for x in range(1,10):

    dt_model = DecisionTreeClassifier(random_state=42,max_depth=x)
    dt_model.fit(X_train,y_train)
    train_score = dt_model.score(X_train,y_train)
    test_score = dt_model.score(X_test,y_test)
    tr_scores.append(train_score)
    te_scores.append(test_score)


plt.title("Bias-Variance")
plt.plot(range(1, 10), tr_scores, label="Train",marker="o")
plt.plot(range(1, 10), te_scores, label="Test",marker="o")
plt.ylabel("Score")
plt.xlabel("Max Depth")       
plt.legend()
plt.grid(linestyle="--")
```

As you can see, we can handle the overfitting problem by changing hyperparameters. If we want to make a good prediction, we should make close **test score & train score** each other. We are looking at the graph, and when the max depth is **3**, we can see clearly that overfitting is minimal.
