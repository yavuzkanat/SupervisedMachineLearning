```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine = load_wine()
```

```python
print(wine.DESCR)
```

```python
df = pd.DataFrame(columns=wine.feature_names,data=wine.data)

df['target'] = wine.target
```

```python
matrix = df.corr()
target_corr = matrix['target']
abs_corr = target_corr.abs()
abs_corr.sort_values(ascending=False)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python

X_train,X_test,y_train,y_test = train_test_split(df[['ash','hue','malic_acid','flavanoids','od280/od315_of_diluted_wines','total_phenols',"alcohol"]],df.target,random_state=42,test_size=0.3)
```

```python
KNN = KNeighborsClassifier()

KNN.fit(X_train,y_train)

print(f"Train Score : {KNN.score(X_train,y_train):.4f}")
print(f"Test Score : {KNN.score(X_test,y_test):.4f}")
```

```python
te_scores = []
tr_scores = []


plt.figure(figsize=(8,5))

for x in range(1,10):

    knn_model = KNeighborsClassifier(n_neighbors=x,weights="uniform")
    knn_model.fit(X_train,y_train)
    train_score = knn_model.score(X_train,y_train)
    test_score = knn_model.score(X_test,y_test)
    tr_scores.append(train_score)
    te_scores.append(test_score)

plt.title("Bias-Variance")
plt.plot(range(1, 10), tr_scores, label="Train Score",marker="o",color="r")
plt.plot(range(1, 10), te_scores, label="Test Score",marker="o",color="g")
plt.ylabel("Score")
plt.xlabel("N_neighbors value")       
plt.legend()
plt.grid(linestyle="--")


print("| Test  | Train  | Test-Train |")
for te,tr in zip(te_scores,tr_scores):
    print(f"|{te:.4f} | {tr:.4f} |  {abs(tr-te):.4f}    |" )

```
